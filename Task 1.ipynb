### Task 1: Replace Hardcoded Prompts
Convert a hardcoded f-string prompt into a reusable LangChain PromptTemplate.

```python
from langchain_core.prompts import PromptTemplate

# 1. Define the template string (No f-strings used here)
template_str = "Explain {topic} in simple terms for beginners"

# 2. Create the PromptTemplate
explain_prompt = PromptTemplate(
    input_variables=["topic"],
    template=template_str
)

# 3. Test the template
topic_input = "Quantum Computing"
result_task1 = explain_prompt.format(topic=topic_input)

print("--- Task 1 Output ---")
print(result_task1)
```

### Task 2: Multi-Input Prompt System
Build a template that accepts multiple inputs: topic, audience, and tone. Run the specified test cases

```python
# 1. Define the multi-input template
multi_template_str = "Explain {topic} for {audience} in a {tone} tone"

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_template_str
)

# 2. Define the test cases provided in the instructions
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("--- Task 2 Output ---")
# 3. Loop through and format the prompt for each test case
for case in test_cases:
    formatted_prompt = multi_prompt.format(**case)
    print(f"Inputs: {case}\nOutput: {formatted_prompt}\n")
```

### Task 3: Prompt Variations Engine
Create three different prompt templates (Teaching, Interview, Storytelling) and run them for the topic "Machine Learning".

```python
# 1. Define templates for different variations
teaching_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} clearly step by step")
interview_template = PromptTemplate(input_variables=["topic"], template="Ask 3 questions about {topic}")
storytelling_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} as a story")

# Store them in a dictionary for easy iteration
variations_engine = {
    "Teaching": teaching_template,
    "Interview": interview_template,
    "Storytelling": storytelling_template
}

# 2. Define the topic
target_topic = "Machine Learning"

print("--- Task 3 Output ---")
# 3. Run the engine for the topic
for style, template in variations_engine.items():
    print(f"[{style} Style]:")
    print(template.format(topic=target_topic))
    print("-" * 20)
```

### Task 4: ChatPrompt Template System
Build a chat-based prompt system utilizing System (behavior) and User (dynamic input) roles

```python
from langchain_core.prompts import ChatPromptTemplate

# 1. Build the chat-based prompt system
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a highly effective {role}."),
    ("user", "Let's talk about {topic}.")
])

# 2. Define roles and the target topic
roles = ["teacher", "interviewer", "motivator"]
chat_topic = "Neural Networks"

print("--- Task 4 Output ---")
# 3. Generate messages for each role
for role in roles:
    messages = chat_template.format_messages(role=role, topic=chat_topic)
    print(f"Role: {role.capitalize()}")
    # Print the resulting message objects to show structure
    for msg in messages:
        print(f"  {msg.type.capitalize()}: {msg.content}")
    print()
```

### Task 5: Input Validation Layer
Create a function to validate audience and tone, assigning default values if the inputs are invalid.

```python
def validate_inputs(audience, tone):
    """Validates audience and tone, assigning defaults if invalid."""

    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Assign defaults if invalid
    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "formal"

    # Optional print statement to show when a default was triggered
    if audience not in valid_audiences:
        print(f"  [Validation Warning] '{audience}' is invalid. Defaulting to 'beginner'.")
    if tone not in valid_tones:
        print(f"  [Validation Warning] '{tone}' is invalid. Defaulting to 'formal'.")

    return final_audience, final_tone

print("--- Task 5 Output ---")
# Test with valid inputs
print("Test 1 (Valid):", validate_inputs("expert", "casual"))

# Test with invalid inputs (should fall back to defaults)
print("Test 2 (Invalid):", validate_inputs("children", "silly"))
```

### Task 6: Prompt Generator App
Combine the template and validation logic into a single app function that generates a structured prompt

```python
def generate_prompt(topic, audience, tone, style):
    # 1. Validate inputs first
    validated_audience, validated_tone = validate_inputs(audience, tone)

    # 2. Define style templates
    # Notice we format it to match the example output provided in the instructions
    styles_dict = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Create interview questions about {topic} for {audience} in a {tone} style",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    # Fallback to teaching if an unknown style is passed
    template_str = styles_dict.get(style, styles_dict["teaching"])

    # 3. Create the prompt template
    app_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template_str
    )

    # 4. Generate the final prompt string
    return app_template.format(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone
    )

print("--- Task 6 Output ---")
# Test matching the example output exactly
example_prompt = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
print("Generated Prompt:\n", example_prompt)
```

### Task 7: Template Reusability Test
Take a single template and run it with 5 different inputs to prove the structure remains the same while the outputs change.

```python
# 1. Define a single reusable template for comparisons
analogy_template = PromptTemplate(
    input_variables=["concept", "analogy"],
    template="Explain the technical concept of {concept} by comparing it to {analogy}."
)

# 2. Define 5 different inputs
reusability_inputs = [
    {"concept": "an API", "analogy": "a restaurant waiter"},
    {"concept": "a Database", "analogy": "a digital filing cabinet"},
    {"concept": "a CPU", "analogy": "the brain of a computer"},
    {"concept": "Encryption", "analogy": "a locked safe with a secret key"},
    {"concept": "Bandwidth", "analogy": "a highway with cars"}
]

print("--- Task 7 Output ---")
# 3. Run the template 5 times
for i, data in enumerate(reusability_inputs, 1):
    print(f"Run {i}:")
    print(analogy_template.format(**data))
    print()
```
